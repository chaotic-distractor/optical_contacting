```python
import csv 
import matplotlib.pyplot as plt
import numpy as np

def extract(filename):
    #get time and temperature from txt file
    results = []
    with open(filename) as tsv:
        for column in zip(*[line for line in csv.reader(tsv, dialect="excel-tab")]):
            columnFloat = []
            for val in list(column):
                columnFloat.append(float(val))
            results.append(columnFloat)
    return results
```

```python
plt.rcParams['figure.figsize'] = [8, 5]

#test 0 - (v2.6) +180 degC in 10 min starting at 21 degC
#top is thermo #1 with Fluke
#bottom is thermo #2 with TPI
test0Data = [extract("PWM_v2.6_180deg_10min_top1.txt"), extract("PWM_v2.6_180deg_10min_bottom2.txt")]

title = "PWM plates - increase 180°C in 10 minutes"
f = plt.figure()
plt.axvline(600, color = 'black', label = '10 minutes')
plt.axhline(200, color ='black', linestyle='-', label = '200°C')

plt.plot([0, 10 * 60], [20, 20+180], c='purple', zorder=0, label = 'linear target')

plt.plot(test0Data[0][0], test0Data[0][1], c='blue', zorder=0, label = 'top (v2.6)')
plt.annotate("139.3", xy=(600, 139.3), xytext=(525, 139.3), size='large', color='red')
plt.plot([600], [139.3], marker='o', markersize=10, color='red')
plt.plot(test0Data[1][0], test0Data[1][1], c='red', zorder=0, label = 'bottom (v2.6)')
plt.annotate("112.3", xy=(600, 112.3), xytext=(615, 102.3), size='large', color='blue')
plt.plot([600], [112.3], marker='o', markersize=10, color='blue')


plt.legend(loc="upper left")
plt.xlabel("Time (sec)")
plt.ylabel("Temperature (°C)")
plt.title(title)
f.savefig(str(title) + ".pdf", bbox_inches='tight')
```
