```python
import csv 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def extract(filename):
    #get time and temperature from txt file
    results = []
    with open(filename) as tsv:
        for column in zip(*[line for line in csv.reader(tsv, dialect="excel-tab")]):
            columnFloat = []
            for val in list(column):
                columnFloat.append(float(val))
            results.append(columnFloat)
    return results
```

```python
plt.rcParams['figure.figsize'] = [8, 5]

test1Cal = [list(range(0, 785)), extract("PWM_v2_test1_calculated_y.txt")[0]]
test1Act = extract("PWM_v2_test1_actual.txt")
test1Av1 = extract("PWM_v2_test1_with_v1.txt")

title = "PWM v2 Test #1 - increase 180°C in 10 minutes"

f = plt.figure()
plt.axvline(600, color = 'blue', label = '10 minutes')
plt.axhline(200, color='cyan', linestyle='-', label = '200°C')
plt.plot(test1Cal[0], test1Cal[1], c='red', zorder=0, label = 'calculated')
plt.plot(test1Av1[0], test1Av1[1], c='grey', zorder=0, label = 'actual (v1)')
plt.plot(test1Act[0], test1Act[1], c='black', zorder=0, label = 'actual (v2)')
plt.legend(loc="upper left")
plt.xlabel("Time (sec)")
plt.ylabel("Temperature (°C)")
plt.title(title)
f.savefig(str(title) + ".pdf", bbox_inches='tight')
```

```python
plt.rcParams['figure.figsize'] = [8, 5]

test2Cal = [list(range(0, 934)), extract("PWM_v2_test2_calculated_y.txt")[0]]
test2Act = extract("PWM_v2_test2_actual.txt")
test2Av1 = extract("PWM_v2_test2_with_v1.txt")

title = "PWM v2 Test #2 - increase 180°C in 7 minutes"

f = plt.figure()
plt.axvline(7 * 60, color = 'blue', label = '7 minutes')
plt.axhline(200, color='cyan', linestyle='-', label = '200°C')
plt.plot(test2Cal[0], test2Cal[1], c='red', zorder=0, label = 'calculated')
plt.plot(test2Av1[0], test2Av1[1], c='grey', zorder=0, label = 'actual (v1)')
plt.plot(test2Act[0], test2Act[1], c='black', zorder=0, label = 'actual (v2)')
plt.legend(loc="upper left")
plt.xlabel("Time (sec)")
plt.ylabel("Temperature (°C)")
plt.title(title)
f.savefig(str(title) + ".pdf", bbox_inches='tight')
```
