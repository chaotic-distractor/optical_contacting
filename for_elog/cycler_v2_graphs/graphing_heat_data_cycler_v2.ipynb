```python
import csv 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def extract(filename):
    #get time and temperature from txt file
    results = []
    with open(filename) as tsv:
        for column in zip(*[line for line in csv.reader(tsv, dialect="excel-tab")]):
            columnFloat = []
            for val in list(column):
                columnFloat.append(float(val))
            results.append(columnFloat)
    return results
```

```python
plt.rcParams['figure.figsize'] = [8, 5]

test1Cal = [list(range(0, 785)), extract("PWM_v2_test1_calculated_y.txt")[0]]
test1Act = extract("PWM_v2_test1_actual.txt")
test1Act2 = extract("PWM_v2.2_test1_actual.txt")
test1Act3 = extract("PWM_v2.3_test1_actual.txt")
test1Av1 = extract("PWM_v2_test1_with_v1.txt")

title = "PWM v2 Test #1 - increase 180°C in 10 minutes"

f = plt.figure()
plt.axvline(600, color = 'blue', label = '10 minutes')
plt.axhline(200, color='cyan', linestyle='-', label = '200°C')
plt.plot(test1Cal[0], test1Cal[1], c='red', zorder=0, label = 'calculated')
plt.plot(test1Av1[0], test1Av1[1], c='whitesmoke', zorder=0, label = 'actual (v1)')
plt.plot(test1Act[0], test1Act[1], c='lightgrey', zorder=0, label = 'actual (v2)')
plt.plot(test1Act2[0], test1Act2[1], c='dimgray', zorder=0, label = 'actual (v2.2)')
plt.plot(test1Act3[0], test1Act3[1], c='black', zorder=0, label = 'actual (v2.3)')

plt.legend(loc="upper left")
plt.xlabel("Time (sec)")
plt.ylabel("Temperature (°C)")
plt.title(title)
f.savefig(str(title) + ".pdf", bbox_inches='tight')
```
